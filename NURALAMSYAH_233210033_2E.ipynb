# CONTOH 1

```python
import matplotlib.pyplot as plt

# Data
gender = [1, 0, 1, 1, 0, 1, 1, 0, 1, 1]
ipk = [3.5, 3.2, 3.8, 2.9, 3.1, 3.6, 2.7, 3.9, 3.3, 3.2]

plt.figure(figsize=(6, 4))

#status 1
pria = [ipk[i] for i in range(len(gender)) if gender[i] == 1]
#status 0
wanita = [ipk[i] for i in range(len(gender)) if gender[i] == 0]

plt.scatter([1]*len(pria), pria, color='r', label='pria (1)')
plt.scatter([0]*len(wanita), wanita, color='y', label='wanita (0)')

# Visualisasi
plt.xlabel('Gender')
plt.ylabel('IPK')
plt.title('Hubungan antara Gender dan IPK')
plt.legend()
plt.xticks([0, 1], ['wanita', 'pria'])
plt.grid(True)
plt.show()
```

# PENJELASAN CONTOH 1

- 'import matplotlib.pyplot as plt' berfungsi untuk mengimpor library  membuat plot dan visual nya.
- gender = (1) mewakili pria (0) mewakili wanita.
- ipk = digunakan untuk memyimpan data ipk setiap gender.
- plt.figure berfungsi untuk membuat plot dengan ukuran yang di buat (satuan inci).
- pada plt.scatter untuk mengimput data atau memberikan identitas pada pria dan wanita.
- lalu pada visualisasi 
    - plt.xlabel('Gender') untuk label variabel x
    - plt.ylabel('IPK') untuk label variabel y
    - plt.title('Hubungan antara Gender dan IPK') untuk judul plot
    - plt.legend() untuk membedakan antara titik pria dan wanita
    - plt.xticks([0, 1], ['wanita', 'pria']) untuk mengatur sumbu x menampilkan wanita dan sumbu y menampilkan pria
    - plt.grid(True) untuk menambah grid pada plot untuk memudahkan pembaca
    - plt.show() menamapilkan hasil plot





# CONTOH 2

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

status_kelulusan = [1, 0, 1, 1, 0, 1, 1, 0, 1, 1]
ipk = [3.5, 3.2, 3.8, 2.9, 3.1, 3.6, 2.7, 3.9, 3.3, 3.2]

# Data Usia dan Ipk

# Histogram
plt.figure(figsize=(6, 4))
sns.histplot(ipk, bins=30, kde=True, color='y')
plt.title('Distribusi Usia berdasarkan umur')
plt.xlabel('Ipk')
plt.ylabel('Status kelulusan')
plt.grid(True)
plt.show()
```

ipk = [3.5, 3.2, 3.8, 2.9, 3.1, 3.6, 2.7, 3.9, 3.3, 3.2] digunakan untuk mengimput data status kelulusan dan ipk 
- pada visualisasi
    - plt.figure(figsize=(6, 4)) digunakan untuk memasukan jenis plot dan ukuran plot dengan satuan inci
    - sns.histplot(ipk, bins=30, kde=True, color='y') digunakan untuk membuat histori nilai dari ipk dan status kelulusan dengan bins dan kurva kepadatan kernel (kde)
    - plt.title('Distribusi Usia berdasarkan umur') digunakan untuk membuat judul
    - plt.xlabel('Ipk') digunakan untuk menambahkan nama lebel variabel x
    - plt.ylabel('Status kelulusan') digunakan untuk menambahkan nama lebel variabel y
    - plt.grid(True) menambahkan grid pada plot agar memudahkan pembaca
    - plt.show() menampilkan hasil plot



# CONTOH 3

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

# Data Usia dan Ipk
usia = [20,22,21,23,20,22,21,24,23,20,22,21]
ipk = [3.5,3.2,3.8,2.9,3.1,3.6,2.7,3.9,3.3,3.2,3.7,3.1]

# Visualisasi
plt.figure(figsize=(6, 4))
sns.regplot(x=usia, y= ipk)
plt.title('Distribusi Usia berdasarkan umur')
plt.xlabel('usia')
plt.ylabel('Ipk')
plt.grid(True)
plt.show()
```

- pada visualisasi
    - plt.figure(figsize=(6, 4)) digunakan untuk memasukan jenis plot dan ukuran plot dengan satuan inci
    - sns.regplot(x=usia, y= ipk) dugunakan untuk membuat jenis plot dengan variabel yang sudah di tentukan
    - plt.title('Distribusi Usia berdasarkan umur') digunakan untuk membuat judul
